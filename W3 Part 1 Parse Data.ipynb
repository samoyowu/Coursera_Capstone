# Part 1: Get Data

```python
import numpy as np
import pandas as pd
```

### Retrieve the URL and parse using BeautifulSoup

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

import requests 
page = requests.get(url)

from bs4 import BeautifulSoup
soup0 = BeautifulSoup(page.content, 'html.parser')
```

### Get Data using for loops

```python
df = pd.DataFrame(columns=range(0,3), index=range(0,181))

row_pt=0
for row in soup0.find_all('tr'):
    col_pt=0
    for item in row.find_all('td'):
        if col_pt<=2:
            df[col_pt][row_pt] = item.text.replace('\n','')
            col_pt+=1
    row_pt+=1


header = []
for h in soup0.find_all('th'):
    header.append(h.text.replace('\n',''))

df.columns = header[0:3]

df.drop(df.index[0], inplace=True)
df.head()
```

### Clean Data: 
Removes cells of which a borough that is Not assigned

```python
df2 = df[df['Borough']!='Not assigned']
df2.head()
```

Check if there is any cells that has a borough but a Not assigned neighborhood (No - as shown in output)
